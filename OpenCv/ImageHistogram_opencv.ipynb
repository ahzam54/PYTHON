### How does image histogram work??
##### * Count number of pixels for each pixel intensity(0-255)

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def grayHistogram():
    root = os.getcwd()
    imgpath = os.path.join(root,'cat.jpg')
    img = cv.imread(imgpath,cv.IMREAD_GRAYSCALE)

    if img is None:
        print(f"Error: Could not load image from {imgpath}")
        return

    plt.figure()
    plt.imshow(img,cmap='gray')

    hist = cv.calcHist([img],[0],None,[256],[0,256])

    plt.figure()
    plt.plot(hist)
    plt.xlabel('bias')
    plt.ylabel('# of pixels')
    plt.show()
```

```python
def colorHistogram():
    root = os.getcwd()
    imgpath = os.path.join(root,'cat.jpg')
    img = cv.imread(imgpath)
    imgRGB = cv.cvtColor(img,cv.COLOR_BGR2RGB)

    plt.figure()
    plt.imshow(imgRGB)

    color = ['b','g','r']
    plt.figure()
    for i in range(len(color)):
        hist = cv.calcHist([imgRGB],[i],None,[256],[0,256])
        plt.plot(hist,color[i])

    plt.xlabel('pixel intensity')
    plt.ylabel('# of pixels')
    plt.show()
```

```python
if __name__ == '__main__':
    grayHistogram()
```

```python
if __name__ == '__main__':
    colorHistogram()
```

```python
def histogramRegion():
    root = os.getcwd()
    imgpath = os.path.join(root,'cat.jpg')
    img = cv.imread(imgpath)
    imgRGB = cv.cvtColor(img,cv.COLOR_BGR2RGB)
    imgRGB = imgRGB[160:230,220:320]

    plt.figure()
    plt.imshow(imgRGB)

    color = ['b','g','r']
    plt.figure()
    for i in range(len(color)):
        hist = cv.calcHist([imgRGB],[i],None,[256],[0,256])
        plt.plot(hist,color[i])

    plt.xlabel('pixel intensity')
    plt.ylabel('# of pixels')
    plt.show()
```

```python
if __name__ == '__main__':
    histogramRegion()
```
